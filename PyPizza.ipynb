 *csv, datetime ve pandas* **Kütüphanelerinin import edilmesi**

```python
import csv
import datetime
import pandas as pd
```

*Klasik, Margarita, Türk ve Plain* **Pizza Sınıflarının oluşturulması**

```python
class Pizza:
    def __init__(self):
        self.description = ''
        self.cost = 0

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.cost


class Classic(Pizza):

    def __init__(self):
        super().__init__()
        self.description = 'klasik pizza'
        self.cost = 20


class Margarita(Pizza):

    def __init__(self):
        super().__init__()
        self.description = 'margarita pizza'
        self.cost = 30


class TurkishPizza(Pizza):

    def __init__(self):
        super().__init__()
        self.description = 'Türk pizza'
        self.cost = 35


class PlainPizza(Pizza):

    def __init__(self):
        super().__init__()
        self.description = 'sade pizza'
        self.cost = 40
```


**Decorator Sınıfların oluşturulması**


```python
class Decorator:
    def get_description(self, pizza):
        return pizza.get_description()

    def get_cost(self, pizza):
        return pizza.get_cost()


class Olive(Decorator):
    decoratorCost = 4

    def get_description(self, pizza):
        pizza.description = 'Zeytinli ' + pizza.get_description()
        super().get_description(pizza)

    def get_cost(self, pizza):
        pizza.cost += Olive.decoratorCost
        super().get_cost(pizza)


class Mushroom(Decorator):
    decoratorCost = 5

    def get_description(self, pizza):
        pizza.description = 'Mantarlı ' + pizza.get_description()
        super().get_description(pizza)

    def get_cost(self, pizza):
        pizza.cost += Mushroom.decoratorCost
        super().get_cost(pizza)


class GoatCheese(Decorator):
    decoratorCost = 8

    def get_description(self, pizza):
        pizza.description = 'Keçi peynirli ' + pizza.get_description()
        super().get_description(pizza)

    def get_cost(self, pizza):
        pizza.cost += GoatCheese.decoratorCost
        super().get_cost(pizza)


class Meat(Decorator):
    decoratorCost = 10

    def get_description(self, pizza):
        pizza.description = 'Etli ' + pizza.get_description()
        super().get_description(pizza)

    def get_cost(self, pizza):
        pizza.cost += Meat.decoratorCost
        super().get_cost(pizza)


class Onion(Decorator):
    decoratorCost = 5

    def get_description(self, pizza):
        pizza.description = 'Soğanlı ' + pizza.get_description()
        super().get_description(pizza)

    def get_cost(self, pizza):
        pizza.cost += Onion.decoratorCost
        super().get_cost(pizza)


class Corn(Decorator):
    decoratorCost = 4

    def get_description(self, pizza):
        pizza.description = 'Mısırlı ' + pizza.get_description()
        super().get_description(pizza)

    def get_cost(self, pizza):
        pizza.cost += Corn.decoratorCost
        super().get_cost(pizza)
```

*Menu.txt* **dosyasının oluşturulması**

```python
with open('menu.txt', 'w') as menu:
    menu.write('* Lütfen Bir Pizza Tabanı Seçiniz:\n')
    menu.write('1: Klasik\n')
    menu.write('2: Margarita\n')
    menu.write('3: TürkPizza\n')
    menu.write('4: Sade Pizza\n\n')
    menu.write('* ve seçeceğiniz malzemeler:\n')
    menu.write('5: Zeytin\n')
    menu.write('6: Mantar\n')
    menu.write('7: Keçi Peyniri\n')
    menu.write('8: Et\n')
    menu.write('9: Soğan\n')
    menu.write('10: Mısır\n\n')
    menu.write('* Teşekkür ederiz!')
```

*main()* **fonksiyonunun oluşturulması**

```python
def main():
    # menu.txt dosyası çağırma ve kullanıcıya okutma
    s = open("menu.txt").read()
    print(s)

    # Kullanıcının seçimi
    pizza_choice = int(input("Pizza seçiminiz (1-4): "))
    while pizza_choice not in [1, 2, 3, 4]:
        pizza_choice = int(input("Geçersiz giriş. Lütfen tekrar deneyin: "))

    # Seçilen pizza
    if pizza_choice == 1:
        pizza = Classic()
    elif pizza_choice == 2:
        pizza = Margarita()
    elif pizza_choice == 3:
        pizza = TurkishPizza()
    else:
        pizza = PlainPizza()

    # Sos seçimi
    durum = 1
    while durum:
        sos_choice = int(input("Sos seçiminiz (5-10): "))
        while sos_choice not in [5, 6, 7, 8, 9, 10]:
            sos_choice = int(input("Geçersiz giriş. Lütfen tekrar deneyin: "))
        if sos_choice == 5:
            sos = Olive()
            sos.get_description(pizza)
            sos.get_cost(pizza)
        if sos_choice == 6:
            sos = Mushroom()
            sos.get_description(pizza)
            sos.get_cost(pizza)
        if sos_choice == 7:
            sos = GoatCheese()
            sos.get_description(pizza)
            sos.get_cost(pizza)
        if sos_choice == 8:
            sos = Meat()
            sos.get_description(pizza)
            sos.get_cost(pizza)
        if sos_choice == 9:
            sos = Onion()
            sos.get_description(pizza)
            sos.get_cost(pizza)
        if sos_choice == 10:
            sos = Corn()
            sos.get_description(pizza)
            sos.get_cost(pizza)
        durum = int(input("Eklemek istediğiniz ek sos var ise '1' yok ise '0' butonuna basınız."))

    print("Sipariş özeti:")
    print(pizza.get_description())
    print("Tutar: ", pizza.get_cost())

    # Kullanıcı bilgileri ve database
    isim = input("Adınız: ")
    tc_no = input("TC kimlik numaranız: ")
    kart_no = input("Kredi kartı numaranız: ")
    sifre = input("Kredi kartı şifresi: ")
    siparis_zamani = datetime.datetime.now().strftime("%d/%m/%Y %H:%M:%S")

    # Siparişi veritabanına kaydedelim.
    df = pd.DataFrame([[pizza.__class__.__name__, isim, tc_no, kart_no, sifre, pizza.get_description(), siparis_zamani]],
                   columns=["Pizza Türü","İsim","TC No","Kredi Kartı No","Kredi Kart Şifre","Sipariş Açıklama","Sipariş Tarihi"])
    with pd.ExcelWriter("Orders_Database.xlsx") as writer:
      df.to_excel(writer, sheet_name='sheet1')
    print("Sipariş bilgileriniz kaydedilmiştir. Bizi tercih ettiğiniz için teşekkür ederiz. Afiyet olsun :)")
```

*main()* **fonksiyonun çalıştırılması**

```python
if __name__ == '__main__':
    main()
```
